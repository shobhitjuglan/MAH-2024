```python
# ####Preprocessing the code
# from sklearn.preprocessing import LabelEncoder

# def preprocess_data(df):
#     # Encode categorical variables
#     le = LabelEncoder()
# #     df = df.dropna()
#     df=df.fillna(df.mean(axis=1), axis=1)
#     for col in df.select_dtypes(include=['object']):
#         df[col] = le.fit_transform(df[col])
#     return df

from sklearn.preprocessing import LabelEncoder

def preprocess_data(df):
    # Encode categorical variables
    le = LabelEncoder()
    
    # Separate numeric and non-numeric columns
    numeric_cols = df.select_dtypes(include=['number']).columns
    non_numeric_cols = df.select_dtypes(include=['object']).columns

    # Fill missing values in numeric columns with the mean of each row
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

    # Encode non-numeric (categorical) columns
    for col in non_numeric_cols:
        df[col] = le.fit_transform(df[col].astype(str))  # Convert to string before encoding

    return df
```

```python
##XGboost classifier

from xgboost import XGBClassifier
id=1

def xgbooster(X_train, X_eval, y_train, type_, eval_ids):
    xgb = XGBClassifier()
    xgb.fit(X_train, y_train)
#     accuracy = xgb.score(X_test, y_test)
    
    # Predict labels for the evaluation dataset (without 'Goal')
    y_eval_pred = xgb.predict(X_eval)
    
    # Create a DataFrame for predictions
    output_df = pd.DataFrame({
        'Id': eval_ids,  # 'Id' column from the evaluation dataset
        'Goal': y_eval_pred  # Predicted 'Goal' labels
    })
    # Save the predictions to a CSV file
    if(type_=="hardware"):
        output_df.to_csv('hardware.csv', index=False)
    elif(type_=="network"):
        output_df.to_csv('network.csv', index=False)
    else:
        output_df.to_csv('os.csv', index=False)
    print("Predictions saved to ",type_, ".csv")

# Create eval_ids for the hardware dataset
eval_ids_h = pd.Series(range(1, len(X_eval_h) + 1))

# Create eval_ids for the network dataset
eval_ids_n = pd.Series(range(len(X_eval_h) + 1, len(X_eval_h) + len(X_eval_n) + 1))

# Create eval_ids for the OS dataset
eval_ids_o = pd.Series(range(len(X_eval_h) + len(X_eval_n) + 1, len(X_eval_h) + len(X_eval_n) + len(X_eval_o) + 1))
    
xgbooster(X_h, X_eval_h, y_h, "hardware", eval_ids_h)
xgbooster(X_n, X_eval_n, y_n, "network", eval_ids_n)
xgbooster(X_o, X_eval_o, y_o, "os", eval_ids_o)
```
